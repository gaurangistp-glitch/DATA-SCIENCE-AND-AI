"is" and "is not" are the identity operators and both are used to check if two values are located on the same part of the memory. Two variables that are equal do not imply that they are identical.

```python
x = 24
y = 20
my_list = [10, 20, 30, 40, 50]

if (x not in my_list):
  print(" x is NOT present in the given list")
else:
  print(" x is present in the given list")

if ( y in my_list):
  print(" y is present in the given list")
else:
  print( " y is NOT prsent in the given list")
```

# ***CONDITIONAL STATEMENTS***

IF STATEMENTS

If statement is used to execute a block of code only when specified condition evaluates to TRUE.

```python
age = 20
if age>= 18 :
  print("Eligible to vote.")
```

SHORT-HAND STATEMENT

Short-Hand statement is used to write if statements in a single line. It is useful when only one statement needs to be executed.

```python
age = 10
if age > 18: print("Eligible to vote.")
```

IF-ELSE STATEMENT

If Else statement is used to execute one block of code when the condition is True and another block when condition is False.

```python
age = 12
if age <= 12:
  print("Travel for free.")
else:
  print("Pay for ticket.")
```

If-elif-else Statement

elif statement is used to check multiple conditions in program. It executes a block of code when its condition evaluates to True after previous conditions evaluate to False.

```python
age = 20

if age <= 12:
  print("CHILD")

elif age <= 19:
  print("TEENAGER")

elif age <= 35:
  print("YOUNG ADULT")

else:
  print("ADULT")
```

NESTED IF-ELSE STATEMENT

Nested if-else statement i an if-else statement placed inside another if or else block. It is used to check conditions within another condition.

```python
age = 70
is_member =True

if age >= 60:
  if is_member:
    print("30% senior discount")

  else:
    print("20% senior discount")

else:
  print("Not eligible for discount")

```

# ***LOOPS***

for LOOP

A for loop iterates over a sequence (list, tuple, string, range etc).

```python
for i in range(1,6):
  print(i)
```

Iterate through list

```python
fruits = ["Apple", "Banana", "Mango", "Watermelon"]

for fruit in fruits :
  print(fruit)
```

WHILE LOOP

A while loop runs as long as condition is True.

```python
count = 1

while count <= 5:
  print(count)
  count += 1
```

Loop Control Statements

A.break

Stops the loop imediately.

```python
for i in range(1, 10):
  if i == 5:
    break
  print(i)
```

B.continue

Skips the current iteration.

```python
for i in range(1, 6):
  if i == 3:
    continue
  print(i)
```

C.pass

Placeholder statement.

```python
for i in range(5):
  pass

print("Loop completed")
```

NESTED LOOP

A loop inside another loop.

```python
for i in range(1, 4):
  for j in range(1, 4):
    print(i, j )
```
